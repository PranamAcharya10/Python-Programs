```python
import numpy as np

def stochastic_gradient_descent(x, y, learning_rate=0.001, epochs=10):
    m_curr = b_curr = 0

    n = len(x)

    for epoch in range(epochs):
        for i in range(n):
            # Pick a random index
            index = np.random.randint(n)
            x_i = x[index]
            y_i = y[index]

            # Calculate the prediction and error
            y_pred = m_curr * x_i + b_curr
            error = y_i - y_pred

            # Update parameters
            m_curr = m_curr + learning_rate * error * x_i
            b_curr = b_curr + learning_rate * error

            # Print updates (optional)
            print(f"Epoch {epoch+1}, Iteration {i+1}: m = {m_curr}, b = {b_curr}")

    return m_curr, b_curr

# Example usage
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

m, b = stochastic_gradient_descent(x, y)
print(f"Final values: m = {m}, b = {b}")
```
