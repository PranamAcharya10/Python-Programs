```python
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[5,7,9,11,13]
```

```python
x=np.array(x)
y=np.array(y)
print(x,y)
```

```python
def gradient_descent(x,y):

  #1 Start with initial value of m and b
  m_curr = b_curr = 0

  #2 Define the number of iterations
  iterations = 10

  #We need n = Length (x)
  n=len(x)

  #We need learning rate, let's start with 0.001
  learning_rate = 0.01

  #3 Simple for loop to iterate 1000 times
  for i in range(iterations):

    #a-Calculate y_predicted using m_curr and b_curr
    y_pred = m_curr*x+b_curr
    print("y_pred = ", y_pred)

    #aa - We'll insert the cost to know how well we perform
    cost = (1/n)*sum([val**2 for val in (y-y_pred)])

    #b - Calculate partial derivatives, m-derivative and b-derivative
    md = -(2/n)*sum(x*(y-y_pred))
    bd = -(2/n)*sum(y-y_pred)

    #c - Update m_curr and b_curr
    m_curr = m_curr - learning_rate*md
    b_curr = b_curr - learning_rate*bd

    #Lets print their values at each iteration
    #Also, print iteration to know what's going on in each iteration
    print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
gradient_descent(x,y)
```
